```python
#makes a data and table
metadata = sqla.MetaData()

stations = sqla.Table(
'stations', metadata,
sqla.Column('number', Integer, primary_key = True),
sqla.Column('name', String(128)),
sqla.Column('address', String(128)),
sqla.Column('pos_lat', Float),
sqla.Column('pos_lng', Float),
sqla.Column('bike_stands', Integer),
sqla.Column('available_bike_stands', Integer),
sqla.Column('available_bike', Integer),
sqla.Column('available_bikes', Integer),
sqla.Column('last_update', DateTime)
)
print(stations.columns)

metadata.create_all(engine)
```
